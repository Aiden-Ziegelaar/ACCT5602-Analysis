```python
import numpy as np

solvency_ratios = {
    'PRU': {
        'Gross Gearing (D/E)' : np.array([5.72, 25.06, 12.97]),
        'Financial Leverage': np.array([1.24, 1.48, 1.33]),
        #TODO: Add other params
        'Net Interest Cover': np.array([2.99, 32.05, 32.00]),
        'Net Debt to Cash Flow': np.array([-0.61, 0.01, -0.21])
    },
    'RRL': {
        'Gross Gearing (D/E)' : np.array([0.3, 4.54, 22.32]),
        'Financial Leverage': np.array([1.33, 1.38, 1.46]),
        #TODO: Add other params
        'Net Interest Cover': np.array([-255.56, 246.82, 110.59]),
        'Net Debt to Cash Flow': np.array([-0.96, -0.61, 0.50])
    }
}


x_axis = ['2019', '2020', '2021']
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()

nic_pru = ax.plot(x_axis, solvency_ratios['PRU']['Net Interest Cover'], color = 'blue', label='PRU Net Interest Cover')
nic_rrl = ax.plot(x_axis, solvency_ratios['RRL']['Net Interest Cover'], color = 'red', label='RRL Net Interest Cover')

plt.xlabel("Year")
plt.ylabel("Net Interest Cover Ratio")

ax2 = ax.twinx()

ndcf_pru = ax2.plot(x_axis, solvency_ratios['PRU']['Net Debt to Cash Flow'], linestyle="dashed", color = 'blue', label='PRU Net Debt to Cash Flow')
ndcf_rrl = ax2.plot(x_axis, solvency_ratios['RRL']['Net Debt to Cash Flow'], linestyle="dashed", color = 'red', label='RRL Net Debt to Cash Flow')


plt.ylabel("Net Debt to Cash Flow Ratio")

plt.title("Long Term Solvency Ratios")

lns = nic_pru+nic_rrl+ndcf_pru+ndcf_rrl
labs = [l.get_label() for l in lns]

plt.legend(lns, labs, bbox_to_anchor=(1.15,1), loc="upper left")

plt.show()
```

```python
fig, ax = plt.subplots()

de_pru = ax.plot(x_axis, solvency_ratios['PRU']['Gross Gearing (D/E)'], color = 'blue', label='PRU Gross Gearing (D/E)')
de_rrl = ax.plot(x_axis, solvency_ratios['RRL']['Gross Gearing (D/E)'], color = 'red', label='RRL Gross Gearing (D/E)')

plt.xlabel("Year")
plt.ylabel("Gross Gearing (D/E) Ratio")

ax2 = ax.twinx()

fl_pru = ax2.plot(x_axis, solvency_ratios['PRU']['Financial Leverage'], linestyle="dashed", color = 'blue', label='PRU Financial Leverage')
fl_rrl = ax2.plot(x_axis, solvency_ratios['RRL']['Financial Leverage'], linestyle="dashed", color = 'red', label='RRL Financial Leverage')

plt.xlabel("Year")
plt.ylabel("Financial Leverage Ratio")

plt.title("Long Term Solvency Ratios")

lns = de_pru+de_rrl+fl_pru+fl_rrl
labs = [l.get_label() for l in lns]

plt.legend(lns, labs, bbox_to_anchor=(1.15,1), loc="upper left")

plt.show()
```
