```python
import numpy as np

liquidity_ratios = {
    'PRU': {
        'Current Ratio' : np.array([3.13,3.3,3.24]),
        'Quick Ratio' :  np.array([1.65,2.24,1.8])
    },
    'RRL': {
        'Current Ratio' :  np.array([3.03,2.74,2.33]),
        'Quick Ratio' :  np.array([2.36,2.00, 1.44])
    }
}
x_axis = ['2019', '2020', '2021']
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()

ax.plot(x_axis, liquidity_ratios['PRU']['Current Ratio'], color = 'blue', label='PRU Current Ratio')
ax.plot(x_axis, liquidity_ratios['PRU']['Quick Ratio'], linestyle="dashed", color = 'blue', label='PRU Quick Ratio')
ax.plot(x_axis, liquidity_ratios['RRL']['Current Ratio'], color = 'red', label='RRL Current Ratio')
ax.plot(x_axis, liquidity_ratios['RRL']['Quick Ratio'], linestyle="dashed", color = 'red', label='RRL Quick Ratio')

plt.xlabel("Year")
plt.ylabel("Ratio")

plt.title("Short Term Liquidity Ratios")

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.show()
```
