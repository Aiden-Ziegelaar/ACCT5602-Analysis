```python
import numpy as np

liquidity_ratios = {
    'PRU': {
        'Current Ratio' : np.array([3.13,3.3,3.24]),
        'Quick Ratio' :  np.array([1.65,2.24,1.8])
    },
    'RRL': {
        'Current Ratio' :  np.array([3.03,2.74,2.33]),
        'Quick Ratio' :  np.array([2.36,2.00, 1.44])
    }
}

liquidity_figures = {
    'PRU': {
        'Current Assets' : np.array([268.47,362.70,403.30]),
        'Current Liabilities' :  np.array([85.83,109.76,124.58])
    },
    'RRL': {
        'Current Assets' : np.array([254.92,277.70,423.52]),
        'Current Liabilities' :  np.array([84.11,101.50,182.13])
    }
}

x_axis = ['2019', '2020', '2021']
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()

ax.plot(x_axis, liquidity_ratios['PRU']['Current Ratio'], color = 'blue', label='PRU Current Ratio')
ax.plot(x_axis, liquidity_ratios['PRU']['Quick Ratio'], linestyle="dashed", color = 'blue', label='PRU Quick Ratio')
ax.plot(x_axis, liquidity_ratios['RRL']['Current Ratio'], color = 'red', label='RRL Current Ratio')
ax.plot(x_axis, liquidity_ratios['RRL']['Quick Ratio'], linestyle="dashed", color = 'red', label='RRL Quick Ratio')

plt.xlabel("Financial Year")
plt.ylabel("Ratio")

plt.title("Short Term Liquidity Ratios")

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.show()
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()

ax.plot(x_axis, liquidity_figures['PRU']['Current Assets'], color = 'blue', label='PRU Current Assets')
ax.plot(x_axis, liquidity_figures['PRU']['Current Liabilities'], linestyle="dashed", color = 'blue', label='PRU Current Liabilities')
ax.plot(x_axis, liquidity_figures['RRL']['Current Assets'], color = 'red', label='RRL Current Assets')
ax.plot(x_axis, liquidity_figures['RRL']['Current Liabilities'], linestyle="dashed", color = 'red', label='RRL Current Liabilities')

plt.xlabel("Financial Year")
plt.ylabel("AUD (m)")

plt.title("Short Term Liquidity Values")

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.show()
```
