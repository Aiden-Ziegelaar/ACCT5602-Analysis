```python
import numpy as np

profitability_ratios = {
    'PRU': {
        'Net Profit Margin' : np.array([1.37,15.99,17.10]),
        'EBITDA Margin': np.array([26.65,41.41,38.25]),
        'NOPLAT Margin': np.array([1.89,16.48,21.05]),
        'ROE': np.array([0.90,10.90,11.27]),
        'ROA': np.array([1.16,7.57,8.68]),
        'ROIC': np.array([1.37,11.12,14.11]),
    },
    'RRL': {
        'Net Profit Margin' : np.array([25.01,26.40,17.85]),
        'EBITDA Margin': np.array([40.48,44.61,34.98]),
        'NOPLAT Margin': np.array([24.91,26.51,18.02]),
        'ROE': np.array([22.77,23.89,9.23]),
        'ROA': np.array([17.21,17.44,6.39]),
        'ROIC': np.array([30.67,29.44,8.70]),
    }
}


x_axis = ['2019', '2020', '2021']
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()

ax.plot(x_axis, profitability_ratios['PRU']['Net Profit Margin'], color = 'blue', label='Net Profit Margin')
ax.plot(x_axis, profitability_ratios['PRU']['EBITDA Margin'], linestyle="dashed", color = 'blue', label='EBITDA Margin')
ax.plot(x_axis, profitability_ratios['PRU']['NOPLAT Margin'], linestyle='dotted', color = 'blue', label='NOPLAT Margin')

plt.xlabel("Year")
plt.ylabel("Ratio (%)")

plt.title("Profitability Ratios - PRU Margin")

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.show()
```

```python
from matplotlib import pyplot as plt

fig, ax = plt.subplots()

ax.plot(x_axis, profitability_ratios['RRL']['Net Profit Margin'], color = 'red', label='Net Profit Margin')
ax.plot(x_axis, profitability_ratios['RRL']['EBITDA Margin'], linestyle="dashed", color = 'red', label='EBITDA Margin')
ax.plot(x_axis, profitability_ratios['RRL']['NOPLAT Margin'], linestyle='dotted', color = 'red', label='NOPLAT Margin')

plt.xlabel("Year")
plt.ylabel("Ratio (%)")

plt.title("Profitability Ratios - RRL Margin")

plt.legend(bbox_to_anchor=(1,1), loc="upper left")

plt.show()
```
